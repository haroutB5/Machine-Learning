## Split data

```python
data_x=data.iloc[:, : 17]
data_y=data[['Revenue']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.3, random_state=42) # 70% training and 30% test

print("Shape of x_train :", X_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", X_test.shape)
print("Shape of y_test :", y_test.shape)
```

## Label encoding  

```python
from sklearn import preprocessing
import pandas as pd
pd.options.mode.chained_assignment = None 
le = preprocessing.LabelEncoder()

le.fit(X_train['Weekend'])
X_train['Weekend']=le.transform(X_train['Weekend'])
X_test['Weekend']=le.transform(X_test['Weekend'])

le.fit(X_train['TrafficType'])
X_train['TrafficType']=le.transform(X_train['TrafficType'])
X_test['TrafficType']=le.transform(X_test['TrafficType'])

le.fit(X_train['VisitorType'])
X_train['VisitorType']=le.transform(X_train['VisitorType'])
X_test['VisitorType']=le.transform(X_test['VisitorType'])

le.fit(X_train['Month'])
X_train['Month']=le.transform(X_train['Month'])
X_test['Month']=le.transform(X_test['Month'])
```

```python
X_train.describe()
```

## Normalization

```python
from sklearn.preprocessing import Normalizer

scaler = Normalizer()
# fit on the training dataset
scaler.fit(X_train.iloc[:, 0 : 10])
# scale the training dataset
X_train.iloc[:, 0 : 10] = scaler.transform(X_train.iloc[:, 0 : 10])
# scale the test dataset
X_test.iloc[:, 0 : 10] = scaler.transform(X_test.iloc[:, 0 : 10])
```

```python
X_train.describe()
```

## See the imbalanced data

```python
sns.countplot(x=y_train['Revenue'])
```

## SMOTE OverSampling 

```python
from imblearn.over_sampling import SMOTE

over = SMOTE()
X, y = over.fit_resample(X_train, y_train)

sns.countplot(x=y['Revenue'])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import auc
from sklearn.metrics import plot_roc_curve
from sklearn.metrics import classification_report
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from scipy.stats import randint
import time
```

## RandomForestClassifier

```python
#the important features selected, run after doing the feature selection at the end
X2=X[['PageValues']]
X_test2=X_test[['PageValues']]
```

```python
from sklearn.ensemble import RandomForestClassifier

#set timer
start_time = time.time()

#create random forest classifier
rfc = RandomForestClassifier(criterion= 'gini', max_depth= 7, max_features= 'sqrt', max_leaf_nodes= 23, min_samples_split= 8, n_estimators= 1000, random_state=42)

#train  random forest classifier
rfc.fit(X2, y.values.ravel())
#rfc.fit(X, y.values.ravel())

#Predict the response for test dataset
y_pred_rfc = rfc.predict(X_test2)
#y_pred_rfc = rfc.predict(X_test)

#show the time this classifier took to perform
print("--- %s seconds ---" % (time.time() - start_time))
```

### results

```python
cr = classification_report(y_test, y_pred_rfc)
print(cr)

pd.DataFrame(confusion_matrix(y_test, y_pred_rfc), 
             columns=['Predicted Negative', 'Predicted Positive'], 
             index=['Actual Negative', 'Actual Positive'])
```

### ROC curve for Random Forrest

```python
plot_roc_curve(rfc, X_test2, y_test)
#plot_roc_curve(rfc, X_test, y_test)

plt.show()
```

### Important features for this Random Forrest

```python
for name, score in sorted(zip(X, rfc.feature_importances_), key = lambda x: x[1], reverse=True):
    print(name, score)
```

### Hyperparameter tuning for the Random Forrest using Random Search

```python
# define space
space = dict(criterion = ['gini', 'entropy'],
         n_estimators=[500, 750, 1000],
         max_features=['auto', 'sqrt', 'log2'],    
         max_depth = randint(1, 15), 
         min_samples_split = randint(2, 15), 
         max_leaf_nodes = randint(5, 30))
   
# define evaluation
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)

# define search
search = RandomizedSearchCV(rfc, space, scoring='accuracy', n_jobs=-1, cv=cv, random_state=42)

# execute search
result = search.fit(X, y.values.ravel())

# summarize result
print('Best Score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```

## Decision tree

```python
#the important features selected, run after doing the feature selection at the end

X3=X[['PageValues', 'Month']]
X_test3=X_test[['PageValues', 'Month']]
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree 

#set timer
start_time = time.time()

# Create Decision Tree classifer object
dtc = DecisionTreeClassifier(criterion= 'gini', max_depth= 9, max_leaf_nodes= 27, min_samples_split= 6)

# Train Decision Tree Classifer
dtc.fit(X3, y)
#dtc.fit(X, y)

#Predict the response for test dataset
y_pred_dtc = dtc.predict(X_test3)
#y_pred_dtc = dtc.predict(X_test)

#show the time this classifier took to perform
print("--- %s seconds ---" % (time.time() - start_time))
```

### results

```python
cr = classification_report(y_test, y_pred_dtc)
print(cr)

pd.DataFrame(confusion_matrix(y_test, y_pred_dtc), 
             columns=['Predicted Negative', 'Predicted Positive'], 
             index=['Actual Negative', 'Actual Positive'])
```

### ROC curve for Decision Tree

```python
plot_roc_curve(dtc, X_test3, y_test)
#plot_roc_curve(dtc, X_test, y_test)

plt.show()
```

### feature importance for Decision Tree

```python
for name, score in sorted(zip(X, dtc.feature_importances_), key = lambda x: x[1], reverse=True):
    print(name, score)
```

### Decision tree hyperparameter optimisation using random search

```python
# define search space
space = dict(criterion = ['gini', 'entropy'],
         max_depth = randint(1, 11), 
         min_samples_split = randint(2, 15), 
         max_leaf_nodes = randint(5, 30))

# define evaluation
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)

# define search
search = RandomizedSearchCV(dtc, space, n_iter=500, scoring='accuracy', n_jobs=-1, cv=cv, random_state=42)

# execute search
result = search.fit(X, y)

# summarize result
print('Best Score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```

## Bagging ensembles


```python
from sklearn.ensemble import BaggingClassifier

#set timer
start_time = time.time()

#create bagging classifier
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(criterion= 'gini', max_depth= 9, max_leaf_nodes= 27, min_samples_split= 6), n_estimators=1000,
    max_samples=100, bootstrap=True, random_state=42)

#train bagging classifier
bag_clf.fit(X, y.values.ravel())

#predict response for test dataset
y_pred_bc = bag_clf.predict(X_test)

#show the time this classifier took to perform
print("--- %s seconds ---" % (time.time() - start_time))
```

### Results

```python
cr = classification_report(y_test, y_pred_bc)
print(cr)

pd.DataFrame(confusion_matrix(y_test, y_pred_bc), 
             columns=['Predicted Negative', 'Predicted Positive'], 
             index=['Actual Negative', 'Actual Positive'])
```

### ROC curve for Bagging ensemble

```python
plot_roc_curve(bag_clf, X_test, y_test)
plt.show()
```

### Bagging ensemble hyperparameter optimisation with grid search

```python
from sklearn.ensemble import BaggingClassifier

model = BaggingClassifier()

#define search space
n_estimators = [10, 100, 500, 1000]
grid = dict(n_estimators=n_estimators)

#define evaluation
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

#define search
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy', error_score=0)

#define result
grid_result = grid_search.fit(X, y.values.ravel())

#print the best result
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
